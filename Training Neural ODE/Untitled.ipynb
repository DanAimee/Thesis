```python
import matplotlib.pyplot as plt

# Dataset
toy_cars = [1, 2, 3, 4, 5]  # Toy car numbers
speeds = [10, 20, 30, 40, 50]  # Speeds of the toy cars in km/h

# Create a line plot
plt.plot(toy_cars, speeds, marker='o', color='skyblue', linestyle='-')

# Add labels and title
plt.xlabel('Toy Car Number')
plt.ylabel('Speed (km/h)')
plt.title('Speed of Toy Cars')

# Set x-axis ticks to match toy car numbers
plt.xticks(toy_cars)

# Show the plot
plt.grid(True)  # Add gridlines for better visualization
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()

```
